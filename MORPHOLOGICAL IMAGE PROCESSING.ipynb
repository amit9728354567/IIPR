```python
import cv2
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("ddownload.png")
```

# BINARY IMAGE

```python
plt.imshow(img)
```

# 3*3 KERNAL

```python
kernal = np.ones((3,3), np.float32)
```

```python
kernal
```

```python
img_erosion = cv2.erode(img, kernal, iterations=1)
img_dilation = cv2.dilate(img, kernal, iterations=1)
```

```python
plt.figure(figsize=(40,20))
plt.subplot(153), plt.imshow(img), plt.title('original')
plt.subplot(151), plt.imshow(img_erosion), plt.title('erosion')
plt.subplot(152), plt.imshow(img_dilation), plt.title('dilation')



plt.show()
```
