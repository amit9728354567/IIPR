```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
img= Image.open('3.jpg')
```

```python
img
```

```python
img.size
img.size[0]
img.size[1]
```

```python
arr=np.asarray(img)
```

```python
max = np.max(arr)
```

```python
for gamma in [0.1, 0.5, 1, 2, 100]:
    for i in range(0,img.size[0]-1):
        for j in range(0,img.size[1]-1):
            pixelColorVals = img.getpixel((i,j));
            c = 255/np.log(1+ max);
            redPixel = c*(pixelColorVals[0]**gamma);
            greenPixel = c*(pixelColorVals[1]**gamma);
            bluePixel = c*(pixelColorVals[2]**gamma);
            img.putpixel((i,j),(redPixel, greenPixel, bluePixel));
    
        
```
