```python
import cv2
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("ddownload.png")
```

# BINARY IMAGE

```python
plt.imshow(img)
```

# CROSS SE

```python
kernal = cv2.getStructuringElement(cv2.MORPH_CROSS, (5,5))
```

```python
img_open = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernal)
```

```python
img_close = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernal)
```

```python
plt.figure(figsize=(40,20))
plt.subplot(153), plt.imshow(img), plt.title('original')
plt.subplot(151), plt.imshow(img_open), plt.title('OPEN')
plt.subplot(152), plt.imshow(img_close), plt.title('CLOSE')



plt.show()
```
