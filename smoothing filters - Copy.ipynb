# Averaging

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("download.jpg",0)
```

```python
plt.imshow(img)
```

```python
m, n = img.shape
```

```python
kernal = np.ones([3,3], dtype=int)
```

```python
kernal = kernal/9
```

```python
new_img = np.ones([m,n])
```

```python
for i in range(1, m-1):
    for j in range(1, n-1):
        temp = img[i-1,j-1]*kernal[0,0]+img[i-1,j]*kernal[0,1]+img[i-1,j+1]*kernal[0,2]+img[i,j-1]*kernal[1,0]+img[i,j]*kernal[1,1]+img[i,j+1]*kernal[1,2]+img[i+1,j-1]*kernal[2,0]+img[i+1,j]*kernal[2,1]+img[i+1,j+1]*kernal[2,2]
        new_img[i,j]=temp
```

```python
plt.imshow(new_img)
```

# Median filter

```python
for i in range(1, m-1):
    for j in range(1, n-1):
        temp = img[i-1,j-1]*kernal[0,0],img[i-1,j]*kernal[0,1],img[i-1,j+1]*kernal[0,2],img[i,j-1]*kernal[1,0],img[i,j]*kernal[1,1],img[i,j+1]*kernal[1,2],img[i+1,j-1]*kernal[2,0],img[i+1,j]*kernal[2,1],img[i+1,j+1]*kernal[2,2]
        a = np.sort(temp)
        median = np.median(a)
        new_img[i,j] = median
```

```python
plt.imshow(new_img)
```
