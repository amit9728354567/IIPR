# Roberts Operator 

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('downloadd.jpg', cv2.COLOR_BGR2GRAY)
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
grayImage = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
kernalx = np.array([[-1,0], [0,1]], dtype=int)
kernaly = np.array([[0,-1],[1,0]], dtype=int)
```

```python
x=cv2.filter2D(grayImage, cv2.CV_16S, kernalx)
y=cv2.filter2D(grayImage, cv2.CV_16S, kernaly)
```

```python
absX = cv2.convertScaleAbs(x)
absY = cv2.convertScaleAbs(y)
```

```python
Roberts = cv2.addWeighted(absX, 0.5, absY, 0.5, 0)
```

```python
titles =['Original Image','Roberts operator']
images =[rgb_img,Roberts]
for i in range(2):
    plt.subplot(1, 2, i+1),plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
plt.show()    
```

# Sobel Operator

```python
sobelkernalx = np.array([[-1,0,1], [-2,0,2],[-1,0,1]], dtype=int)
sobelkernaly = np.array([[-1,-2,-1],[0,0,0],[1,2,1]], dtype=int)
```

```python
sx=cv2.filter2D(grayImage, cv2.CV_16S, sobelkernalx)
sy=cv2.filter2D(grayImage, cv2.CV_16S, sobelkernaly)
```

```python
sabsX = cv2.convertScaleAbs(sx)
sabsY = cv2.convertScaleAbs(sy)
```

```python
Sobels = cv2.addWeighted(sabsX, 0.5, sabsY, 0.5, 0)
```

```python
titles =['Original Image','Sobels operator']
images =[rgb_img,Sobels]
for i in range(2):
    plt.subplot(1, 2, i+1),plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
plt.show()    
```

# prewitt operator

```python
pkernalx = np.array([[1,0,-1], [1,0,-1],[1,0,-1]], dtype=int)
pkernaly = np.array([[1,1,1],[0,0,0],[-1,-1,-1]], dtype=int)
```

```python
px=cv2.filter2D(grayImage, cv2.CV_16S, pkernalx)
py=cv2.filter2D(grayImage, cv2.CV_16S, pkernaly)
```

```python
pabsX = cv2.convertScaleAbs(px)
pabsY = cv2.convertScaleAbs(py)
```

```python
Prewitt = cv2.addWeighted(pabsX, 0.5, pabsY, 0.5, 0)
```

```python
titles =['Original Image','Prewitt operator']
images =[rgb_img,Prewitt]
for i in range(2):
    plt.subplot(1, 2, i+1),plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
plt.show()    
```
