# 1D

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.array([1,2,8,9])
x
```

# DFT

```python
dftx=np.fft.fft(x)
dftx
```

```python
magx=np.abs(dftx)
magx
```

```python
angx=np.angle(dftx)
angx
```

# INVERSE DFT 

```python
idftx=np.fft.ifft(x)
idftx
```

# 2D

```python
x1=np.ones([4,4])
dftx1=np.fft.fft2(x1)
dftx1
```

```python
magx1=np.abs(dftx1)
magx1
```

```python
angx1=np.angle(dftx1)
angx1
```

```python
image1=np.zeros((400,400), dtype='uint8')
a=cv2.rectangle(image1, (100,100), (250,250), 255, -1)
```

```python
plt.imshow(a)
```

```python
dfta=np.fft.fft2(a)
dftas=np.fft.fftshift(dfta)
maga=np.abs(dftas)
anga=np.angle(dftas)
```

```python
back=np.fft.ifft2(dftas)
back=np.abs(back)
```

```python
plt.figure(figsize=(20,10))
plt.subplot(151), plt.imshow(a,'gray'), plt.title('original')
plt.subplot(152), plt.imshow(np.log(np.abs(dfta)),'gray'), plt.title('DFT')
plt.subplot(153), plt.imshow(np.log(maga), 'gray'), plt.title('magnitude spectrum of shifted center')
plt.subplot(154), plt.imshow(anga,'gray'), plt.title('angle')
plt.subplot(155), plt.imshow(back,'gray'), plt.title('inverse fft')
plt.show()
```
