# Convolution


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("download.jpg")
```

```python
kernal = np.ones((5,5), np.float32)/25
dst = cv2.filter2D(img, -1, kernal)

```

```python
plt.subplot(121), plt.imshow(img), plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(dst), plt.title('Averaging')
plt.xticks([]), plt.yticks([])
plt.show()
```

# Blur

```python

blur = cv2.blur(img, (5,5))
```

```python
plt.subplot(121), plt.imshow(img), plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(blur), plt.title('Averaging')
plt.xticks([]), plt.yticks([])
plt.show()
```

# Median blur

```python
mblur = cv2.medianBlur(img, 89)
```

```python
plt.subplot(121), plt.imshow(img), plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(mblur), plt.title('Averaging')
plt.xticks([]), plt.yticks([])
plt.show()
```

# Boxfilter

```python
box = cv2.boxFilter(img, -1, (5,5), normalize= True)
```

```python
plt.subplot(121), plt.imshow(img), plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(box), plt.title('Averaging')
plt.xticks([]), plt.yticks([])
plt.show()
```
